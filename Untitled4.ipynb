```python
def simpleMA(length, option, ch):

    callback.historical_Data=ch

    result = 0
    Sum = 0
    
    time.sleep(5)
    i = len(callback.historical_Data)-1-1
    n=length
    c = 0
    if option == 1:
        #print(callback.historical_Data)
        while i > (len(callback.historical_Data)-1-1-n):
            if callback.historical_Data[i][3]>900 and callback.historical_Data[i][4]>900:
                print(callback.historical_Data[i][1])
             
                Sum = Sum +(callback.historical_Data[i][3]+callback.historical_Data[i][4])/2
                c = c + 1
            i = i-1
        result = Sum/c
        return result
    if option == 0:
        while i > (len(callback.historical_Data)-1-1-n):
            if  callback.historical_Data[i][4]>900:
                Sum = Sum +(callback.historical_Data[i][4])
                c = c+1
            i = i-1
        result = Sum/c
        return result
    if option == 2:
        while i > (len(callback.historical_Data)-1-1-n):
            if callback.historical_Data[i][3]>900:
                c=c+1
                Sum = Sum +(callback.historical_Data[i][3])
            i = i-1
        result = Sum/c
        return result
```
